```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib as mpl

rcParams['figure.figsize'] = 15, 6
rcParams['font.size'] = 20
```

```python
import numpy as np
import math as m
```

```python
def Rx(theta):
    return np.matrix([[1, 0, 0],
                     [0, m.cos(theta), -m.sin(theta)],
                     [0, m.sin(theta), m.cos(theta)]])


def Ry(theta):
    return np.matrix([[m.cos(theta), 0, m.sin(theta)],
                     [0, 1, 0],
                     [-m.sin(theta), 0, m.cos(theta)]])


def Rz(theta):
    return np.matrix([[m.cos(theta), -m.sin(theta), 0],
                     [m.sin(theta), m.cos(theta), 0],
                     [0, 0, 1]])
```

```python
def RT():
    e11
    e12
    e13
    e11
    e12
    e13
    
    return np.matrix([[m.cos(theta), -m.sin(theta), 0],
                      
                     [m.sin(theta), m.cos(theta), 0],
                      
                     [0, 0, 1]])
```

```python
# In radiants
angleInDegreePhi = 90.0
angleInDegreeTheta = 0.0
angleInDegreePsi = 0.0
phi = np.radians(angleInDegreePhi)
theta = np.radians(angleInDegreeTheta)
psi = np.radians(angleInDegreePsi)

print("phi =", phi)
print("theta  =", theta)
print("psi =", psi)
  
# Get the rotational matrix  
R = Rz(psi) * Ry(theta) * Rx(phi)
print(np.round(R, decimals=2))
```

```python
import sys
tol = sys.float_info.epsilon * 10

if abs(R.item(0, 0)) < tol and abs(R.item(1, 0)) < tol:
    eul1 = 0
    eul2 = m.atan2(-R.item(2, 0), R.item(0, 0))
    eul3 = m.atan2(-R.item(1, 2), R.item(1, 1))
else:
    eul1 = m.atan2(R.item(1, 0), R.item(0, 0))
    sp = m.sin(eul1)
    cp = m.cos(eul1)
    eul2 = m.atan2(-R.item(2, 0), cp*R.item(0, 0)+sp*R.item(1, 0))
    eul3 = m.atan2(sp*R.item(0, 2)-cp*R.item(1, 2),
                   cp*R.item(1, 1)-sp*R.item(0, 1))

print("phi =", eul1)
print("theta  =", eul2)
print("psi =", eul3)
```

# Rotate in space

```python
v1 = np.array([[1],[0],[0]])

angleInDegreePhi = 90.0
angleInDegreeTheta = 0.0
angleInDegreePsi = 0.0
phi = np.radians(angleInDegreePhi)
theta = np.radians(angleInDegreeTheta)
psi = np.radians(angleInDegreePsi)

print("phi =", phi)
print("theta  =", theta)
print("psi =", psi)
  
# Get the rotational matrix  
R = Rz(psi) * Ry(theta) * Rx(phi)

# rotate in space
v2 = R * v1
```
